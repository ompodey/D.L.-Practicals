```python
from collections import deque
from concurrent.futures import ThreadPoolExecutor
from typing import Dict, List, Set
dir(typing)
```

```python
def parallelbfs(graph : Dict[int, List[int]], start_node: int) -> List[int]:
    visited=set()
    queue = deque([start_node])
    result=[]
    
    
      
    def processel(current_el : int, visited : Set[int],  queue : deque):
        if current_el not in visited:
            visited.add(current_el)
            result.append(current_el)
            
            for sider in reversed (graph [current_el]):
                if sider not in visited:
                    queue.append(sider)
    
    
    
    with ThreadPoolExecutor() as executor:
        
        while queue:
            current_el = queue.popleft()
        
            executor.submit(processel, current_el, visited , queue)    
        

    return result
```

```python
graph = {
    1:[2,3],
    2:[1,4,5],
    3:[1,6],
    4:[2],
    5:[2],
    6:[3]
}

res = parallelbfs(graph,1)
print(res)
```
